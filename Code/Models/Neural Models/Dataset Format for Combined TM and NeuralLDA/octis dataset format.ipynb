```python
# # Assign split percentages
train_percent = 0.8
val_percent = 0.1

# Determine the split for each row
total_rows = len(df)
train_rows = int(train_percent * total_rows)
val_rows = int(val_percent * total_rows)

# Create the 'Split' column and assign split labels
split_labels = ['train'] * train_rows + ['val'] * val_rows + ['test'] * (total_rows - train_rows - val_rows)
np.random.shuffle(split_labels)  # Shuffle the labels randomly

df['Split'] = split_labels

df.head()
```

```python
df['Label'] = actual_Label
df.head()
```

```python
def remove_nonbreaking_space(text):
    return re.sub(r'\xa0', ' ', text)

df['Clean_Data'] = df['Clean_Data'].apply(remove_nonbreaking_space)
```

```python
import nltk
stopwords=pd.read_csv('/content/drive/MyDrive/stopwords.txt',names=['List'])
# stopwords['List']

stopwords_ur=[]
for li in stopwords['List']:
  stopwords_ur.append(li)
print(len(stopwords_ur))
print(stopwords_ur)
```

```python
def remove_stopwords(text: str, stopwords_ur):
    return " ".join(word for word in text.split() if word not in stopwords_ur)
```

```python
df['Clean_Data'] = df['Clean_Data'].apply(lambda text: remove_stopwords(text, stopwords_ur))
```

```python
for document in df['Clean_Data'].iloc[:2]:
    print(document)
    print('---')
```

```python

from gensim.corpora import Dictionary
from typing import List

def create_vocabulary(documents, save_to: str = None) -> List[str]:

    dataset=[[word for word in str(document).split()] for document in documents]
    vocab = Dictionary(documents=dataset)
    word_list = [key for key in vocab.token2id.keys() if len(key) > 1]
    print(len(word_list))
    if save_to:
        with open(f"{save_to}/vocabulary.txt", 'w') as txt:
            for word in word_list:
                txt.write(word + '\n')
    return word_list
```

```python
create_vocabulary(documents, save_to="sample_data")
```
